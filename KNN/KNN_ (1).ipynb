```python
x=df.drop(columns=["income"])
x
```

```python
y=df.income
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.33,random_state=42,stratify=y)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler()
X=sc.fit_transform(x)
X
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
#to check the prediction is right
accuracy_score(pred,y_test)
```

```python
#to check the model is overfit or underfit
trainpred=knn.predict(x_train)
```

```python
accuracy_score(trainpred,y_train)
```

```python
#parameter
for k in range(1,100,2):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    predict=knn.predict(x_test)
    print("k= {} Accuracy {}".format(k,accuracy_score(y_test,predict)))
```

```python
from sklearn.model_selection import GridSearchCV
knn=KNeighborsClassifier(n_jobs=-1)
param_grid={
    "n_neighbors":[23,25],
    "weights":["uniform","distance"],
    "algorithm":["auto","ball_tree", "kd_tree","brute"],

}
gsv=GridSearchCV(knn,param_grid,n_jobs=-1,refit=True)
gsv.fit(x_train,y_train)
prediction=gsv.predict(x_test)
print(gsv.best_params_)
print(gsv.best_score_)
print(accuracy_score(y_test,prediction))
```
